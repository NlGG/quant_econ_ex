```julia
using PyPlot

function generate_data(n)
    epsilon_values = Array(Float64, n)
    for i = 1:n
        epsilon_values[i] = randn()
    end
    return epsilon_values
end

ts_length = 100
data = generate_data(ts_length)
plot(data, "b-")
```

```julia
using PyPlot
using Distributions

function plot_histogram(distribution, n)
    epsilon_values = rand(distribution, n)  # n draws from distribution
    plt[:hist](epsilon_values)
end

lp = Laplace()
plot_histogram(lp, 500)
```

```julia
function my_lin_interp(z)
    function func(x)
        y = 2*x*z
        return y
    end

    return func
end
```

```julia
f = my_lin_interp(10)
```

```julia
f(2) #2*x*10
```

```julia
grid = [1, 2]
vals = [2, 0]
f = my_lin_interp(grid, vals)

f(1.25)
```

```julia
using Grid
using PyPlot

f(x) = 2 .* cos(6x) .+ sin(14x) .+ 2.5
c_grid = 0:.2:1

Af = CoordInterpGrid(c_grid, f(c_grid), BCnil, InterpLinear)

f_grid = linspace(0, 1, 150)

fig = figure()
ax = fig[:add_subplot](111)

ax[:plot](f_grid, f(f_grid), "b-", lw=2, alpha=0.8, label="true function")
ax[:plot](f_grid, Af[f_grid], "g-", lw=2, alpha=0.8,
          label="linear approximation")

ax[:vlines]([c_grid], [c_grid] * 0, f(c_grid), linestyle="dashed", alpha=0.5)
ax[:legend](loc="upper center")
```

```julia
function MC(n)
    x = Zeros(n)
    y = Zeros(n)
    for i in 1:n
        x[i] = 2*rand()-1
        y[i] = 2*rand()-1
    end
    plot(x, y)
    n = 0
    for i in 1:n
        if x[i]^2 + y[i]^2 < 1
            n+=1
        end
    end
end
    
    
```

```julia
MC(100)
```
